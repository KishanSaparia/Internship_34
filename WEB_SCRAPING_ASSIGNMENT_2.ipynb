# Q1
Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")

# search on google is https://www.naukri.com/
driver.get("https://www.naukri.com/") 

# enter the value for designation Data Analyst
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Analyst")

# enter the value for Location is Bangalore
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
location.send_keys('Bangalore')

# for click on search button
search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
time.sleep(3) # this time is require to open site


# for Job title
title=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
job_title=[i.text for i in title[0:10]]
# print(job_title)
# print('\n')

# for job location
location=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
job_location=[i.text for i in location[0:10]]
# print(job_location)
# print('\n')

# for company
company=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
company_name=[i.text for i in company[0:10]]
# print(company_name)
# print('\n')

# for experience required
experience=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
experience_required=[i.text for i in experience[0:10]]
# print(experience_required)
# print('\n')



# creating Data Frame
df=pd.DataFrame({'job-title':job_title,'job-location':job_location,'company_name':company_name,'experience_required':experience_required})
df
```

# Q4
Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

Brand

ProductDescription

Price

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")

# search on google is https://www.flipkart.com/
driver.get("https://www.flipkart.com/") 
# time.sleep(15)


# for click on login close button
search=driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')  
search.click()
# time.sleep(3) # this time is require to open site

# enter the value sunglasses
Search_on=driver.find_element(By.CLASS_NAME,"_3704LK")
Search_on.send_keys("sunglasses")


#click on search button
search=driver.find_element(By.CLASS_NAME,'L0Z3Pu')
search.click()
time.sleep(3)

Brand=[]
Product=[]
Price=[]

#fetch all data in nested loop
for x in range(0,3):
    brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    productDescription=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    
    for i in brand:
        Brand.append(i.text)
    
    for i in productDescription:
        Product.append(i.text)
        
    for i in price:
        Price.append(i.text)
    
    click_on_next=driver.find_element(By.XPATH,"//span[text()='Next']")
    click_on_next.click()
    time.sleep(3)

#Creating of Data Frame    
df=pd.DataFrame({
    "Brand_name":Brand[0:100],
    "ProductDescription":Product[0:100],
    "Price":Price[0:100]
})
df
```

# Q7
Go to webpage https://www.amazon.in/

Enter “Laptop” in the search field and then click the search icon 

After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:

Title

Ratings

Price

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")

# search on google is https://www.amazon.in/
driver.get("https://www.amazon.in/") 


# enter the value laptop
Search_on=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
Search_on.send_keys("Laptop")

# for click on search_button
search_button=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')  
search_button.click()
time.sleep(3)



# for click on intel i7
search_intel=driver.find_element(By.XPATH,"//span[text()='Intel Core i7']")  
search_intel.click()
time.sleep(3)

#fetch data on page
title=driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
rating=driver.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]')
price=driver.find_elements(By.XPATH,'//span[contains(@class,"a-price-whole")]')

Title=[i.text for i in title[0:10]]
Rating=[i.text for i in rating[0:5]]
Price=[i.text for i in price[0:5]]
Rating1=[i.text for i in rating[13:18]]
Price1=[i.text for i in price[13:18]]
Rating=Rating+Rating1
Price=Price+Price1

#creating Data Frame
df=pd.DataFrame({
    'Title':Title,
    'Rating':Rating,
    'Price':Price
})
df
```

# Q8
Write a python program to scrape data for Top 1000 Quotes of All Time.

Than scrap a) Quote b) Author c) Type Of Quotes

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")

# search on google is https://www.azquotes.com/
driver.get("https://www.azquotes.com/") 



# for click on Top Quotes
search_Top_Quotes=driver.find_element(By.XPATH,"//a[text()='Top Quotes']")  
search_Top_Quotes.click()
time.sleep(3)




quote=[]
author=[]
type_of_quotes=[]
#fetch all data in nested loop 
for x in range(0,10):
    Quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    Author=driver.find_elements(By.XPATH,'//div[@class="author"]')
    Type_Of_Quotes=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    
    for i in Quote:
        quote.append(i.text)
    for i in Author:
        author.append(i.text)
    for i in Type_Of_Quotes:
        type_of_quotes.append(i.text)
    x=x+1
    if x>=10:
        break
    else:
        Next_page=driver.find_element(By.XPATH,'//li[@class="next"]')  
        Next_page.click()
        time.sleep(3)
        
#creating Data Frame
df=pd.DataFrame({
    'Quote':quote,
    'Author':author,
    'Type_of_quotes':type_of_quotes
})
df
```

# Q10
Write a python program to display list of 50 Most expensive cars in the world (i.e.
Car name ,Description and Price) from https://www.motor1.com/

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")

# search on google is https://www.motor1.com/
driver.get("https://www.motor1.com/")
time.sleep(3)

#Click to open dropbox
Click_d=driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[1]/div")
Click_d.click()
time.sleep(1)

#click on features
Click_f=driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[5]/button")
Click_f.click()
time.sleep(1)

#click on list
Click_l=driver.find_element(By.XPATH,'//a[text()="Lists"]')
Click_l.click()
time.sleep(1)

#click on 50 Most Expensive Cars In The World
Click_50=driver.find_element(By.XPATH,'//a[text()="50 Most Expensive Cars In The World"]')
Click_50.click()
time.sleep(2)

#Car Name
name=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
name=[i.text for i in name[0:50]]
# print(len(name))

#Car price
price=driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
price=[i.text for i in price[0:50]]
# print(len(price))


#Car Descriptipons
description=driver.find_elements(By.XPATH,'//div[@class="postContent content-wrapper"]/div/p')
Description=[i.text for i in description]
x=Description[4:]

des=[]
for i in range(0,99,2):
    if i%2==0:
        des.append(x[i])
# print(des)
# print(len(des))

#Create Data Frame
df=pd.DataFrame({
    "Car_Name":name,
    "Description":des,
    "Price":price
})
df
```
